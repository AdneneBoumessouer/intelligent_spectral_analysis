```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import utils
```

```python
plot_params = {'xlabel':"Raman Shift $[cm^{-1}]$", 
               'ylabel':"Intensity", 
               'grid':'both', 
               'figsize':(18,8), 
               'legend':None}
```

```python
lambda0 = 522
rs_min = 0
rs_max = 3000
```

## Load raw data

```python
df = utils.read_data("data/4.0_january/pentandiol.txt")
```

```python
# split data according to setup (power, integration time)
df_24_5 = df.filter(regex='x|24mW_5s').filter(regex='x|Pentan')
df_24_10 = df.filter(regex='x|24mW_10s').filter(regex='x|Pentan')
df_34_5 = df.filter(regex='x|34mW_5s').filter(regex='x|Pentan')
df_34_10 = df.filter(regex='x|34mW_10s').filter(regex='x|Pentan')
```

```python
# transform wavelength axis into ramanshift
df_24_5 = utils.wavelength2ramanshift(df_24_5, ref=lambda0, x='x')
df_24_10 = utils.wavelength2ramanshift(df_24_10, ref=lambda0, x='x')
df_34_5 = utils.wavelength2ramanshift(df_34_5, ref=lambda0, x='x')
df_34_10 = utils.wavelength2ramanshift(df_34_10, ref=lambda0, x='x')
```

```python
ax = df_24_5.plot(x='x', title='24mW_5s, all concentrations', **plot_params)
ax.axvline(x=rs_min, ymin=0, ymax=1, color='red')
ax.axvline(x=rs_max, ymin=0, ymax=1, color='red')
```

```python
ax = df_24_10.plot(x='x', title='24mW_10s, all concentrations', **plot_params)
ax.axvline(x=rs_min, ymin=0, ymax=1, color='red')
ax.axvline(x=rs_max, ymin=0, ymax=1, color='red')
```

```python
ax = df_34_5.plot(x='x', title='34mW_5s, all concentrations', **plot_params)
ax.axvline(x=rs_min, ymin=0, ymax=1, color='red')
ax.axvline(x=rs_max, ymin=0, ymax=1, color='red')
```

```python
ax = df_34_10.plot(x='x', title='34mW_10s, all concentrations', **plot_params)
ax.axvline(x=rs_min, ymin=0, ymax=1, color='red')
ax.axvline(x=rs_max, ymin=0, ymax=1, color='red')
```
